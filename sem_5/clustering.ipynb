```r
library(e1071)
library(rpart)
library(ggplot2)
library(readr)
```

```r
movies <- read.csv("u.item", sep="|")
```

```r
colnames(movies)<- c("movie_id" , "movie_title" , "release_date" , "video_release_date" , "IMDb_URL" , "unknown" , "Action" , "Adventure" , "Animation" , "Children's" , "Comedy" , "Crime" , "Documentary" , "Drama" , "Fantasy" , "Film-Noir" , "Horror" , "Musical" , "Mystery" , "Romance" , "Sci-Fi" , "Thriller" , "War" , "Western")
```

```r
head(movies)
```

```r
wss <- 0

for (i in 1:18) {
  km.out <- kmeans(movies[,c(-1:-6)], centers = i, nstart = 30, iter.max = 1000)
  
  wss[i] <- km.out$tot.withinss
}
 
```

```r
plot(1:18, wss, type = "b", 
     xlab = "Number of Clusters", 
     ylab = "Within groups sum of squares")
```

```r
new_movies <- movies[,c(-1:-6)]
head(new_movies)
moviesCluster<- kmeans(new_movies, centers = 90, nstart = 10, iter.max = 5000)
moviesCluster
```

```r
# add cluster with movie id into new data frame
movies_by_cluster<- data.frame(cbind(movies$movie_id, clusterNum = moviesCluster$cluster))
names(movies_by_cluster)<-c("movie_id", "cluster")
head(movies_by_cluster)
```

```r
# users
users = read.csv('u.user', sep='|')
colnames(users)<-c('user_id','age','gender','occupation','zip_code')
head(users)
```

```r
# all data
full_data = read.csv('u.data', sep='\t')
colnames(full_data)<-c("user_id","movie_id","rating" ,"timestamp")
head(full_data)
```

```r
# merge full data with clusterd movies
total_data <- merge(full_data, movies_by_cluster, by=c("movie_id"))
```

```r
tail(total_data)
```

```r
wss <- 0

for (i in 1:10) {
  km.out <- kmeans(total_data, centers = i, nstart = 10, iter.max = 1000)
  
  wss[i] <- km.out$tot.withinss
}
 

```

```r
plot(1:18, wss, type = "b", 
     xlab = "Number of Clusters", 
     ylab = "Within groups sum of squares")
```
